**You have created a dataset `df`. Plot `c vs d` as a scatter plot to recreate the plot below. Try to match the color and size of the data points. You should be able to match the size of the figure as well!**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
df = pd.DataFrame(np.random.rand(1000,4),columns=['a', 'b', 'c','d'])
%matplotlib inline
df.plot.scatter(x='c',y='d', color='green', figsize=(7,3))
```

**Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?**<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
df = pd.DataFrame(np.random.rand(1000,4),columns=['a', 'b', 'c','d'])
%matplotlib inline
df.plot.scatter(x='a',y='b', s=df['d']*60, color='green', figsize=(5,3))
```

**Recreate the plot below**

```python
df.plot.hexbin(x='a',y='b', gridsize=20, cmap='coolwarm',  figsize=(6,4))
```

** Create a kde plot of the 'd' column **

```python
df['d'].plot.kde(linewidth=1.0, figsize=(4,3))
```
